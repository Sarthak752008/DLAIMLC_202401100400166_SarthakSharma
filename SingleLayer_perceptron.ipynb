```python
X = [75, 65, 80]          # inputs
W = [4, 3, 5]             # weights
b = -0.1                 # bias
lr = 0.01                 # learning rate
y = 1                     # target output (Pass)

# Step 2: Net input (z)
z = (W[0]*X[0]) + (W[1]*X[1]) + (W[2]*X[2]) + b
print("z before learning:", z)

# Step 3: Activation function
y_pred = 1 if z >= 0 else 0
print("Prediction before learning:", y_pred)

# Step 4: Error
error = y - y_pred
print("Error:", error)

# Step 5: Weight update (if error != 0)
if error != 0:
    for i in range(len(W)):
        W[i] = W[i] + lr * error * X[i]

# Step 6: Bias update
b = b + lr * error

print("Updated Weights:", W)
print("Updated Bias:", b)

# Step 7: Recalculate z
z_new = (W[0]*X[0]) + (W[1]*X[1]) + (W[2]*X[2]) + b
y_new = 1 if z_new >= 0 else 0

print("z after learning:", z_new)
print("Prediction after learning:", y_new)
```
